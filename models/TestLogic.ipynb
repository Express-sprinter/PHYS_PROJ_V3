```python
import random
from collections import defaultdict

# --- Generate points ---
def make_new_point(min_val=-100, max_val=100):
    x = random.randint(min_val, max_val)
    y = random.randint(min_val, max_val)
    z = random.randint(min_val, max_val)
    return (x, y, z)

# --- Partition points into cubes ---
def partition_points(points, cube_size):
    grid = defaultdict(list)
    for x, y, z in points:
        cube_index = (
            int(x // cube_size),
            int(y // cube_size),
            int(z // cube_size)
        )
        grid[str(cube_index)].append((x, y, z))
    return dict(grid)  # regular dict

# --- Main runner ---
def run(num_points=100, min_val=-100, max_val=100, cube_size=10):
    points = [make_new_point(min_val, max_val) for _ in range(num_points)]
    grid_data = partition_points(points, cube_size)
    return grid_data, points

# Example usage
NUM_POINTS = 500
CUBE_SIZE = 10
MIN_VAL = -50
MAX_VAL = 50

data, points = run(NUM_POINTS, MIN_VAL, MAX_VAL, CUBE_SIZE)

# data is now a Python dictionary with your cube keys and points

def pretty_print_dict(d):
    for key, values in d.items():
        print(f"{key}:")
        for v in values:
            print(f"    {v}")  # indented
        print()  # blank line between keys
        
        
#pretty_print_dict(data)




def workWithData(inputData, p, cz):

    start = random.choice(p)
    end = start
    while end == start:
        end = random.choice(p)
    
    vStart = [float(x) for x in start.strip("()").split(" ")]
    vEnd = [float(x) for x in end.strip("()").split(" ")]
    
    directionalVector = []
    cubeDirectionalVector = []
    for idx,itm in enumerate(start):
        D = end[idx]-itm
        directionalVector.append(D)
        cubeDirectionalVector.append(int(D // cube_size))
    
        
    return directionalVector, cubeDirectionalVector

```

```python
def voxel_line(x0, y0, z0, x1, y1, z1):
    x, y, z = int(x0), int(y0), int(z0)
    x1, y1, z1 = int(x1), int(y1), int(z1)

    dx = abs(x1 - x0)
    dy = abs(y1 - y0)
    dz = abs(z1 - z0)

    step_x = 1 if x1 > x0 else -1
    step_y = 1 if y1 > y0 else -1
    step_z = 1 if z1 > z0 else -1

    # Prevent division by zero
    tMaxX = ((x + (step_x > 0)) - x0) / (x1 - x0) if dx != 0 else float('inf')
    tMaxY = ((y + (step_y > 0)) - y0) / (y1 - y0) if dy != 0 else float('inf')
    tMaxZ = ((z + (step_z > 0)) - z0) / (z1 - z0) if dz != 0 else float('inf')

    tDeltaX = abs(1 / (x1 - x0)) if dx != 0 else float('inf')
    tDeltaY = abs(1 / (y1 - y0)) if dy != 0 else float('inf')
    tDeltaZ = abs(1 / (z1 - z0)) if dz != 0 else float('inf')

    voxels = [(x, y, z)]

    while (int(x) != int(x1)) or (int(y) != int(y1)) or (int(z) != int(z1)):
        # Advance to next voxel boundary
        if tMaxX < tMaxY:
            if tMaxX < tMaxZ:
                x += step_x
                tMaxX += tDeltaX
            else:
                z += step_z
                tMaxZ += tDeltaZ
        else:
            if tMaxY < tMaxZ:
                y += step_y
                tMaxY += tDeltaY
            else:
                z += step_z
                tMaxZ += tDeltaZ

        voxels.append((x, y, z))

    return voxels
```
